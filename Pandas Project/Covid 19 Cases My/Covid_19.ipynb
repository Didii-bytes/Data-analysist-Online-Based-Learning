```python
import pandas as pd
import matplotlib.pyplot as plt

#load csv file 
df = pd.read_csv("cases_malaysia.csv")

#see first 5 row
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Convert date colummn to datetime
df['date']=pd.to_datetime(df['date'])
```

```python
# Total number of cases
print("Total cases:", df['cases_new'].sum())

# Average cases per day
print("Average cases per day:", df['cases_new'].mean())

# Day with highest cases
print("Day with highest cases:", df[df['cases_new'] == df['cases_new'].max()])
```

```python
# Plot daily cases
plt.figure(figsize=(14,6))
plt.plot(df['date'], df['cases_new'], label='Daily New Cases')
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.title("Daily COVID-19 Cases in Malaysia")
plt.legend()
plt.show()
```
